# Modifying columns

Strip removes spaces and taps on this column after that make it upper case

```python
df['Branch'] = df['Branch'].str.strip().str.upper()
df['Branch']
```

Strip removes spaces and taps on this column after that make the first letter is capital remaining are small case

```python
df['City'] = df['City'].str.strip().str.title()
df['City']
```

Check null values if it is available in dataset

```python
null_values = df[df['unit_price'].isnull()]
print(null_values)
```

Filling missing values

```python
for i in df.columns:
    if df[i].dtype==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

# Drop Duplicate Invoice_id

```python
df.drop_duplicates(subset=['invoice_id'], inplace=True)
```

```python
df.shape
```

converts the date column to a datetime type and errors='coerce' it will be replaced with NaT (Not a Time) instead of raising an error.

```python
df['date'] = pd.to_datetime(df['date'], format="%d/%m/%y", errors='coerce')
```

```python
df['date']
```

It performs a backward fill operation on the date column

```python
df['date'] = df['date'].bfill()
```

```python
df['date']
```

checks how many missing or NaN values in the date column

```python
df['date'].isnull().sum()
```

Converts the values in the time column to datetime objects.

```python

df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df['unit_price'] = df['unit_price'].str.replace('$','').astype(float)
df['unit_price']
```

Total Revenue for each transaction

```python
df['Revenue'] = df['unit_price']*df['quantity']
```
